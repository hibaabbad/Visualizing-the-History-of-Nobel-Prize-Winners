The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

![](Nobel_Prize.png)

The Nobel Foundation has made a dataset available of all prize winners from the outset of the awards from 1901 to 2023. The dataset used in this project is from the Nobel Prize API and is available in the `nobel.csv` file in the `data` folder.

In this project, you'll get a chance to explore and answer several questions related to this prizewinning data. And we encourage you then to explore further questions that you're interested in!

```python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np

# Start coding here!
df= pd.read_csv('data/nobel.csv')
print(df.columns)
```

```python
top_gender=df['sex'].value_counts()
top_gender='Male'
top_country=df['birth_country'].value_counts()
top_country='United States of America'
```

```python
df['Us_born']=df['birth_country']=='United States of America'
df['decade']=(df['year'] // 10) * 10
```

```python
decade_stats = df.groupby('decade').agg(
    total_winners=('year', 'count'),
    us_born_winners=('Us_born', 'sum')
).reset_index()

# Calculate the ratio of US-born winners to total winners
decade_stats['us_born_ratio'] = decade_stats['us_born_winners'] / decade_stats['total_winners']

# Identify the decade with the highest ratio
max_decade_usa = decade_stats.loc[decade_stats['us_born_ratio'].idxmax()]
max_decade_usa=int(max_decade_usa[0])
print(max_decade_usa)
```

```python
df['female_win']=df['sex']=='Female'
decade_prop = df.groupby(['decade','category']).agg(
    total_winners=('year', 'count'),
    female_winners=('female_win', 'sum')
).reset_index()

decade_prop['female'] = decade_prop['female_winners'] / decade_prop['total_winners']
max_female = decade_prop.loc[decade_prop['female'].idxmax()]
max_female_dict={max_female[0] : max_female[1]}
print(max_female_dict)
```

```python
female_winners = df[df['female_win'] == True]
min_year = female_winners['year'].min()
first_woman = female_winners[female_winners['year'] == min_year].iloc[0]
first_woman_name = first_woman['full_name']
first_woman_category = first_woman['category']

print(first_woman_name, first_woman_category)
```

```python
count_indiv = df.groupby('full_name')['prize'].count()
repeat_winners = count_indiv[count_indiv > 1]
repeat_list = repeat_winners.index.tolist()
print(repeat_list)
```
